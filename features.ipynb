# Выделение геометрических и текстурных признаков

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from PIL import Image,ImageOps
import cv2
import os

images_path = "images"

```

## Предобработка исходного изображения

```python
for filename in os.listdir(images_path):
    img = np.asarray(Image.open(f"{images_path}/{filename}").convert("L"))
    std = np.zeros(img.shape)
    mean = np.mean(img)
    for i in range(len(img)):
        for j in range(len(img[i])):
            if img[i][j]:
                var0 = np.var(img)**0.5
                var1 = np.var(img[i][j]-mean)**0.5
                std[i][j] = var1/var0 + mean
                print(var0,var1)
    cv2.imwrite(f"std/{filename}", std)
    break
```

### Гистограммные признаки

```python
# Гистограмма для признаков mean, var, skew, kurt, perc01, perc10, perc50, perc90, perc 99
def get_hist(img: np.ndarray) -> list:
    hist = [0]* 256
    rows, cols = img.shape
    for i in range(rows):
        for j in range(cols):
            hist[img[i][j]] += 1
    hist = [color/(rows*cols) for color in hist]
    return hist

#Mean среднее значение
def getMean(hist: list) -> float:
  mean=0
  for i in range (len(hist)):
    mean+=(i+1)*hist[i]
  return mean

#Variance дисперсия
def getVariance(hist: list,mean: float) -> float:
  variance=0
  for i in range (len(hist)):
    variance+=pow((i+1)-mean,2)*hist[i]
  return variance


#Skewness коэффициент ассиметрии
def getSkewness(hist: list,mean: float,variance: float) -> float:
  skewness=0
  for i in range (len(hist)):
    skewness+=pow((i+1)-mean,3)*hist[i]
  return skewness/pow(variance,1.5)


#Kurtosis коэффициент эксцесса
def getKurtosis(hist: list,mean: float,variance: float) -> float:
  kurtosis=0
  for i in range (len(hist)):
    kurtosis+=pow((i+1)-mean,4)*hist[i]
  return (kurtosis/pow(variance,2)-3)


#Perc процентные характеристики гистограммы (1,10,50,90,99 процентов) (количество элементов гистограммы, сумма которых меньше заданного процента)
def getPerc(hist: list,val: int) -> float:
  temp=0
  perc=0
  while temp<val/100:
    temp+=hist[perc]
    perc+=1
  return perc

img = np.asarray(Image.open("images/frame0.png").convert("L"))
hist = get_hist(img)
mean = getMean(hist)
var = getVariance(hist, mean)
skew = getSkewness(hist, mean, var)
krut = getKurtosis(hist, mean, var)
print(hist)
print(mean)
print(var)
print(skew)
print(krut)
for p in [1,10,50,90,99]:
  print(getPerc(hist,p))
```
